```python
import pandas as pd
df = pd.read_excel("Book1.xlsx")
df.head()
```

```python
df.info()
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.barplot(x="Player", y="Runs", data=df)
plt.xticks(rotation=45)
plt.title("Runs by Cricket Players")
plt.show()
```

```python
sns.scatterplot(x="StrikeRate", y="Runs", data=df)
plt.title("Strike Rate vs Runs")
plt.show()
```

```python
df_sorted = df.sort_values("Runs", ascending=False)
df_sorted[["Player", "Runs"]]
```
